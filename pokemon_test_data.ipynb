```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
pokemon_name = []
all_pokemon = 'https://serebii.net/pokedex-sv/'
result = requests.get(all_pokemon)
soup = bs(result.content, 'lxml')

tables=soup.find_all('table')
second_table = tables[1]

options = second_table.find_all('option')
pokemon_names = [option.get_text(strip=True) for option in options]

pokemon_list = []
for option in options:
    text = option.get_text(strip=True)
    number, name = text.split(' ', 1) 
    pokemon_list.append((number, name))

for number, name in pokemon_list:
    print(name.lstrip())
```

# Analisando a página do bubasaur

```python
pokemon_website = 'https://serebii.net/pokedex-sv/bulbasaur/'
result = requests.get(pokemon_website)
content = result.text
soup = bs(content, 'lxml')
```

```python
content = soup.find_all('table', class_='dextable')
content = content[1].find_all('td', class_='fooinfo')
content
```

```python
df = pd.DataFrame()
```

```python
name = content[0].get_text()
df['name'] = [name]
df
```

```python
gender = content[3].get_text()
male = gender.split(':')[1].split('%')[0]
female = gender.split(':')[2].split('%')[0]
df["percent_male"] = [male]
df["percent_female"] = [female]
df
```

```python
content_type = soup.find_all('table', class_='dextable')
content_type = content_type[1].find_all('td', class_='cen')
img_tag = content_type[0]
img_tag = img_tag.find_all('img')

pokemon_type = []

for i in img_tag:
    pokemon_type.append(i['alt'].split('-')[0])  

if len(pokemon_type) == 1:
    pokemon_type.append("null")

df["type_1"] = pokemon_type[0] 
df["type_2"] = pokemon_type[1]

df
```

```python
classification = content[4].get_text()
df["classification"] = classification
df
```

```python
height = content[5].get_text()
height = height.split('\t')[-1]
df["height"] = height
df
```

```python
weight = content[6].get_text().split("\t")[-1]
df["weight"] = weight
df
```

```python
capture_rate = content[7].get_text()
df["capture_rate"] = capture_rate
df
```
